```python
def get_inconsistenty_flag(x):
    vio_all = [x.iloc[2], x.iloc[3], x.iloc[4], x.iloc[5]]
    cur_min = min(vio_all)
    min_idx = vio_all.index(cur_min)
    
    if min_idx == 0: # G0 Y0
        if x.iloc[0] == 0 and x.iloc[1] == 0:
            return 0
        else:
            return 1
    elif min_idx == 1: # G0 Y1
        if x.iloc[0] == 1 and x.iloc[1] == 0:
            return 0
        else:
            return 1
    elif min_idx == 2: # G1 Y0
        if x.iloc[0] == 0 and x.iloc[1] == 1:
            return 0
        else:
            return 1
    else:
#         min_idx == 2: # G1 Y1
        if x.iloc[0] == 1 and x.iloc[1] == 1:
            return 0
        else:
            return 1
```

```python
seeds = [1, 12345, 6, 2211, 15, 88, 121, 433, 500, 1121, 50, 583, 5278, 100000, 0xbeef, 0xcafe, 0xdead, 7777, 100, 923]
# seeds = [88, 121, 433, 500, 1121, 50, 583, 5278, 100000, 0xbeef, 0xcafe, 0xdead, 7777, 100, 923]

models = ['LR', 'TR']

datasets = ['lsac', 'cardio', 'bank', 'meps16', 'credit', 'ACSE', 'ACSP', 'ACSH', 'ACSM', 'ACSI']
```

```python
all_diff_df = pd.DataFrame()
eval_path = repo_dir + 'eval/'
for data_name in datasets:
    eval_df = pd.read_csv('{}n-sim-both-all-{}.csv'.format(eval_path, data_name))
    all_diff_df = pd.concat([all_diff_df, eval_df])
```

```python
all_diff_df.head()
```

```python
# first_cols = ['size', '# of attributes (numer./ categ.)', 'minority group U', 'population of U', '% of positive labels in U', 'predictive task']

rows = [['model', 'case'] + datasets]


err_df = pd.DataFrame(columns=['model', 'case'] + datasets)
cases_names = ['0 to 1', '1 to 0']
  
for model_i in ['lr', 'tr']:
    for case_i in cases_names:
        cur_row = [model_i, case_i]
        for data_name in datasets:
            cur_ratios = []
            for seed_i in seeds:
                cur_df = all_diff_df.query('data=="{}" and model=="{}" and sim=="{}" and seed =={} and method=="SEP"'.format(data_name, model_i, case_i, seed_i))
#                 print(cur_df)
                cur_ratios.append(cur_df['sim_n'].values[0]/cur_df['group_n'].values[0])
            
#             cur_ratios = [round(x/y, 3) for x, y in zip(cur_df['sim_n'], cur_df['group_n'])]
#             print(cur_ratios)
            cur_row.append('{}%'.format(round(np.mean(cur_ratios)*100, 1)))
            
        err_df.loc[err_df.shape[0]] = cur_row
        rows.append(cur_row)
```
